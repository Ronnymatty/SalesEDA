## This command is used to import all relevant libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
files = [file for file in os.listdir("./Sales")]

for file in files :
    print(file)
```

```python
all_sales = pd.DataFrame()
for file in files :
    df = pd.read_csv("./Sales/"+file, sep=",")
    all_sales = pd.concat([all_sales,df])
    

all_sales
```

```python
all_sales.isnull().sum()
```

```python
all_sales=all_sales.dropna(how="all")
```

```python
all_sales.isnull().sum()
```

```python
all_sales.dtypes
```

```python
all_sales["Quantity Ordered"]=pd.to_numeric(all_sales["Quantity Ordered"])
```

```python
all_sales[all_sales["Quantity Ordered"]=="Quantity Ordered"]
```

```python
all_sales=all_sales[all_sales["Quantity Ordered"]!="Quantity Ordered"]
```

```python
all_sales[all_sales["Quantity Ordered"]=="Quantity Ordered"]
```

```python
all_sales["Quantity Ordered"]=pd.to_numeric(all_sales["Quantity Ordered"])
```

```python
all_sales["Price Each"]=pd.to_numeric(all_sales["Price Each"])
```

```python
all_sales.dtypes
```

```python
all_sales["Month"]=all_sales["Order Date"].str[0:2]
```

```python
all_sales
```

```python
all_sales["Month"]=all_sales["Month"].astype("int32")
```

```python
all_sales.dtypes
```

```python
all_sales["Sales"]=all_sales["Quantity Ordered"]*all_sales["Price Each"]
```

```python
all_sales
```

```python
result=all_sales.groupby("Month").sum()
```

```python
result
```

```python
plt.bar(result.index,result["Sales"])
plt.xlabel("Month")
plt.ylabel("Total Sales per Month")
plt.title("Monthly Sales")
plt.xticks(result.index)
plt.ticklabel_format(style="plain",axis="y")
plt.show()
```

```python
result1=all_sales.groupby("Product").sum()
```

```python
result1
```

```python
plt.barh(result1.index,result1["Quantity Ordered"])
plt.ylabel("Product")
plt.xlabel("Quantity Ordered by Product")
plt.title("Quantity of Product Ordered")
#plt.xticks(result.index,rotation=90)
#plt.ticklabel_format(style="plain",axis="y")
plt.show()
```

```python
result2=all_sales.groupby("Product").sum().sort_values(by="Quantity Ordered")
```

```python
plt.barh(result2.index,result2["Quantity Ordered"])
plt.ylabel("Product")
plt.xlabel("Quantity Ordered by Product")
plt.title("Quantity of Product Ordered")
#plt.xticks(result.index,rotation=90)
#plt.ticklabel_format(style="plain",axis="y")
plt.show()
```

```python
result3=all_sales.groupby("Product").sum().sort_values(by="Quantity Ordered",ascending=False)
```

```python
plt.barh(result3.index,result3["Quantity Ordered"])
plt.ylabel("Product")
plt.xlabel("Quantity Ordered by Product")
plt.title("Quantity of Product Ordered")
#plt.xticks(result.index,rotation=90)
#plt.ticklabel_format(style="plain",axis="y")
plt.show()
```
